```python
## Collaborative Filtering Based Recommender System
```

```python
# Selecting only those users which have rated more than 200 books in the application

x=ratings['user-id'].value_counts()>200
y=x[x].index
```

```python
y
```

```python
# Keeping the information of only those users which are in y

ratings=ratings[ratings['user-id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.head()
```

```python
# Finding the number of ratings of each book

number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'number_of_ratings'},inplace=True)
```

```python
number_rating.head()
```

```python
# Merging the number_rating dataframe with ratings_with_books dataframe

final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating.head()
```

```python
final_rating=final_rating[final_rating['number_of_ratings']>=50]
```

```python
final_rating.sample(10)
```

```python
# Removing those cases if a user has rated a particular book multiple times

final_rating.drop_duplicates(['user-id','title'],inplace=True)
```

```python
# Creating a pivot table

book_pivot=final_rating.pivot_table(columns='user-id',index='title',values='rating')
```

```python
book_pivot
```

```python
# Filling the NaN values with 0

book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
# Creating a sparse matrix for rows

from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
# Finding the recommendations of the book id 237 which is Harry Potter and the Chamber of Secrets (Book 2)

distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
# Distances of the suggested books are

distances
```

```python
# Index of suggested books are

suggestions
```

```python
# Names of suggested books are

for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
# Creating a function that would take a book's name and recommend the related books

def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6) 

    for i in range(len(suggestions)):
        if not i:
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```
